```python
from geopy.geocoders import ArcGIS
```

```python
nom=ArcGIS()
```

```python
nom.geocode("Dieselstraße 12, 51103 Köln")
```

```python
data = pandas.read_csv("supermarkets/supermarkets.csv")
data
```

```python
data['Address'] = data['Address'] + ", " + data['City'] + ", " + data['State'] + "," + data['Country']
data
```

```python
data["Coordinates"] = data["Address"].apply(nom.geocode)
data
```

```python
data["Latitude"] = data["Coordinates"].apply(lambda x: x.latitude)
```

```python
data["Longitude"] = data["Coordinates"].apply(lambda x: x.longitude)
```

```python
data.drop("Coordinates", axis=1)
data
```
