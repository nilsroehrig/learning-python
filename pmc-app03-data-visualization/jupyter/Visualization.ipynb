```python
import pandas
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt

data = pandas.read_csv("../materials/reviews.csv", parse_dates=["Timestamp"])
data.head()
```

### Rating average/count by day

```python

data["Day"] = data["Timestamp"].dt.date
day_average = data.groupby(["Day"]).mean()
day_average.head()
```

```python
plt.figure(figsize=(25, 3))
plt.plot(day_average.index, day_average["Rating"])
```

### Average rating by week

```python
data["Week"] = data["Timestamp"].dt.strftime('%Y-%U')
week_avg = data.groupby(["Week"]).mean()

plt.figure(figsize=(25,3))
plt.plot(week_avg.index, week_avg["Rating"])
```

### Average rating by month

```python
data["Week"] = data["Timestamp"].dt.strftime('%Y-%m')
month_avg = data.groupby(["Week"]).mean()

plt.figure(figsize=(25,3))
plt.plot(month_avg.index, month_avg["Rating"])
```

### Average rating by month by course

```python
data["Month"] = data["Timestamp"].dt.strftime("%Y-%m")
month_avg_course = data.groupby(["Month", "Course Name"])["Rating"].mean().unstack()
# month_avg_course = data.groupby(["Month", "Course Name"])["Rating"].count().unstack()

month_avg_course.plot(figsize=(25,8))
```

### What day are people the happiest?

```python
data["Weekday"] = data["Timestamp"].dt.strftime("%A")
data["Daynumber"] = data["Timestamp"].dt.strftime("%w")

wd_avg = data.groupby(["Weekday", "Daynumber"]).mean()
wd_avg = wd_avg.sort_values("Daynumber")

plt.figure(figsize=(15,3))
plt.plot(wd_avg.index.get_level_values(0), wd_avg["Rating"])
```

### Number of ratings by course

```python
share = data.groupby(["Course Name"])["Rating"].count()


plt.pie(share, labels=share.index)
```
